```python
import pandas as pd
```

```python
bonus_points = pd.read_csv("Bonus Points AF 2013 - 2016.csv")
school_reference_codes = pd.read_csv("schoolcodes.csv")
```

```python
school_reference_codes['CountyCode'] = school_reference_codes['CountyCode'].astype("int").astype("str")
school_reference_codes['DistrictCode'] = school_reference_codes['DistrictCode'].fillna(value = (school_reference_codes['CountyCode'] + school_reference_codes['DistrictKey']))
```
